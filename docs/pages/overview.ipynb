
**puzzle-widget** is a Jupyter widget where students drag scrambled code
lines into the correct order. Tag a cell `%%puzzle <result>`; every line
below it becomes a draggable row. Drag a row (or focus it and press the
arrow keys) to move it up or down -- the widget re-runs the current order
after every move and shows a checkmark once it produces `<result>`.

## How it works

`%%puzzle <result>` parses `<result>` as a Python literal
(`ast.literal_eval`) and splits the cell into non-blank, whitespace-stripped
lines -- each becomes one draggable row, in the order written. After every
reorder, `puzzle_widget.checker.run_puzzle` executes the *current* order in
a fresh, isolated namespace and reads off the value of the last line (a bare
expression, or a simple assignment's target); the widget shows a success
banner once that value equals `<result>`.

A puzzle's lines must be independent, non-indented, single-line statements
-- no multi-line `if`/`for`/`def`/... blocks -- so that *every* ordering at
least parses; only the value produced should depend on getting the order
right.
